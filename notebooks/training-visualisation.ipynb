```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
parameters = {"train_loss": dict(title="Train loss", y= "BCE+DICE Loss"), "val_loss": dict(title="Validation loss", y= "BCE+DICE Loss"), "val_dice": dict(title="Test Dice", y= "DICE Coefficient"), "val_iou": dict(title="Test IoU", y= "IoU Coefficient") }
```

```python
csv_logs_dir = os.path.join(os.curdir, "..", "csv-logs")
trials_names = os.listdir(csv_logs_dir)
trials_names
```

```python
trials_1 = ['unet-raw', 'unet-autoencoder-1-encoder-unfrozen', 'unet-autoencoder-1']
name_maps_1 = {"unet-raw": "Random initialization", "unet-autoencoder-1": "Autoencoder weights (Forzen encoder)", "unet-autoencoder-1-encoder-unfrozen": "Autoencoder weights"}
fig_1_dir = os.path.join(os.curdir, "..", "logs-visualisation", "figure-1")

for param in parameters.keys():
    plt.figure(figsize=(5, 5))
    for trial in trials_1[::-1]:
        path = os.path.join(csv_logs_dir, trial, f"{param}.csv")
        data = pd.read_csv(path)
        plt.plot(data["Value"], label=name_maps_1[trial])
        plt.ylabel(ylabel=parameters[param]["y"])
        plt.xlabel(xlabel="Epoch")
        plt.legend()
    
    plt.savefig(fname=os.path.join(fig_1_dir, f"{param}.png"))
```

```python
trials_2 = ['unet-raw', 'unet-encoder-only', 'unet-double-autoencoder']
name_maps_2 = {"unet-raw": "Random initialization", "unet-encoder-only": "Pretrained encoder only", "unet-double-autoencoder": "Pretrained encoder and decoder"}
fig_2_dir = os.path.join(os.curdir, "..", "logs-visualisation", "figure-2")

for param in parameters.keys():
    plt.figure(figsize=(5, 5))
    for trial in trials_2[::-1]:
        path = os.path.join(csv_logs_dir, trial, f"{param}.csv")
        data = pd.read_csv(path)
        plt.plot(data["Value"], label=name_maps_2[trial])
        plt.ylabel(ylabel=parameters[param]["y"])
        plt.xlabel(xlabel="Epoch")
        plt.legend()
    
    plt.savefig(fname=os.path.join(fig_2_dir, f"{param}.png"))
```
